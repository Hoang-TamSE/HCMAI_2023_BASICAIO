```python
import shutil
import os

# Define the source and destination folders
src = "/home/aivn2020/HCMCAI/code/HCMAI_2023/SenmaticSearchCLIP/Keyframes_L01/keyframes" # The folder that contains the subfolders with images
dst = "/home/aivn2020/HCMCAI/code/HCMAI_2023/SenmaticSearchCLIP/image" # The folder where you want to copy the images

# Loop through the subfolders in the source folder
for subfolder in os.listdir(src):
# Join the subfolder name with the source folder path
    subfolder_path = os.path.join(src, subfolder)
    # Check if the subfolder is a directory
    if os.path.isdir(subfolder_path):
    # Loop through the files in the subfolder
        for file in os.listdir(subfolder_path):
    # Join the file name with the subfolder path
            file_path = os.path.join(subfolder_path, file)
            # Check if the file is an image (you can use other criteria here)
            if file.endswith(".jpg"):
            # Copy the image to the destination folder
                shutil.copy(file_path, dst)
```

```python
import os
import shutil

def move_all_images(source_folder, destination_folder):
    # Iterate over the subfolders in the source folder
    for root, dirs, files in os.walk(source_folder):
        # Iterate over the files in each subfolder
        for file in files:
            # Check if the file is an image (you can modify the condition based on your requirements)
            if file.endswith(('.jpg', '.jpeg', '.png', '.gif')):
                # Construct the source and destination file paths
                source_path = os.path.join(root, file)
                destination_path = os.path.join(destination_folder, file)
                # Move the image file to the destination folder
                shutil.move(source_path, destination_path)
                print(f"Moved {file} to {destination_folder}")

# Specify the parent folder containing the subfolders with images
parent_folder = "/Downloads/Keyframes_L01/keyframes"

# Specify the destination folder where all images will be moved
destination_folder = "/home/aivn2020/HCMCAI/code/HCMAI_2023/SenmaticSearchCLIP/image"

# Call the function to move all the images
move_all_images(parent_folder, destination_folder)
```
